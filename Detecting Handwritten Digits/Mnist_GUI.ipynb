```python
import keras
import matplotlib.pyplot as plt
from matplotlib import image
from skimage.color import rgb2gray
import numpy as np
import gradio as gr
```

```python
mymodel = keras.models.load_model('/content/drive/MyDrive/Mnist_Trained/Mnist_CNN.h5')
# mymodel.summary()
```

```python
image = gr.inputs.Image(shape=(28,28))
# label = gr.outputs.Number()
def myfunc(img):
    greyimg = rgb2gray(img).reshape(-1,28,28,1)
    x=str(np.argmax(mymodel.predict(greyimg)))
    return x
```

```python
gr.Interface(myfunc, inputs=image, outputs='number').launch()
```
